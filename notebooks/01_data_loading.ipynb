```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
prices = pd.read_csv("../data/raw/prices.csv")

# Parse datetime
prices["datetime"] = pd.to_datetime(prices["datetime"])
prices = prices.set_index("datetime")

# Plot
plt.figure(figsize=(10, 4))
plt.plot(prices.index, prices["price"])
plt.title("Hourly Electricity Prices (Â£/MWh)")
plt.ylabel("Price")
plt.xlabel("Time")
plt.tight_layout()
plt.show()
```
