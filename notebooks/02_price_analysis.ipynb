```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
prices = pd.read_csv("../data/raw/prices.csv")
prices["datetime"] = pd.to_datetime(prices["datetime"])
prices = prices.set_index("datetime")

# Basic statistics
prices.describe()
```

```python
# Add hour column
prices["hour"] = prices.index.hour

# Average price by hour
hourly_avg = prices.groupby("hour")["price"].mean()

plt.figure(figsize=(10, 4))
hourly_avg.plot(kind="bar")
plt.title("Average Electricity Price by Hour")
plt.ylabel("Price (Â£/MWh)")
plt.xlabel("Hour of Day")
plt.tight_layout()
plt.show()
```

Prices are lowest overnight and peak during morning and evening demand periods. This intraday volatility creates arbitrage opportunities for flexible assets such as battery storage.
